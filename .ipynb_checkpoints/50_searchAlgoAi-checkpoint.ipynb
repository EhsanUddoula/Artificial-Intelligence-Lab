```python
import heapq
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import deque
```

### Functions to generate maze
Will run the searching algorithms in this problem

```python
# Maze definition
TERRAIN_COSTS = {
    '.': 1,  # flat
    '^': 3,  # hill
    '~': 5,  # swamp
    '#': -1,  # obstacle
    'S': 0,   # start (no cost)
    'G': 0    # goal (no cost)
}

# Create a 15x15 maze with different terrains
def generate_maze():
    maze = [
        ['S', '.', '.', '^', '.', '#', '.', '.', '.', '#', '.', '^', '.', '.', '.'],
        ['.', '#', '^', '.', '.', '.', '#', '^', '.', '.', '.', '#', '.', '~', '.'],
        ['^', '.', '#', '.', '~', '.', '.', '.', '#', '.', '~', '.', '.', '.', '#'],
        ['.', '.', '~', '#', '.', '.', '#', '.', '.', '.', '.', '#', '^', '.', '.'],
        ['#', '.', '.', '.', '#', '^', '.', '.', '~', '.', '.', '.', '.', '#', '.'],
        ['.', '#', '.', '~', '.', '.', '#', '.', '.', '#', '.', '.', '~', '.', '.'],
        ['.', '.', '#', '.', '.', '~', '.', '.', '^', '.', '#', '.', '.', '.', '^'],
        ['^', '.', '.', '#', '.', '.', '#', '.', '.', '.', '.', '#', '.', '.', '.'],
        ['.', '#', '.', '.', '~', '.', '.', '#', '^', '.', '.', '.', '#', '.', '~'],
        ['~', '.', '#', '.', '.', '#', '.', '.', '.', '#', '.', '^', '.', '.', '.'],
        ['.', '.', '.', '#', '^', '.', '.', '~', '.', '.', '#', '.', '.', '#', '.'],
        ['#', '.', '~', '.', '.', '#', '.', '.', '.', '^', '.', '.', '#', '.', '.'],
        ['.', '#', '.', '.', '#', '.', '~', '.', '#', '.', '.', '~', '.', '.', '#'],
        ['.', '.', '#', '^', '.', '.', '.', '#', '.', '.', '^', '.', '.', '#', '.'],
        ['.', '~', '.', '.', '#', '.', '^', '.', '.', '#', '.', '.', '~', '.', 'G']
    ]
    return maze
```

```python
# Visualization functions
def visualize_maze(maze, path=None):
    plt.figure(figsize=(10, 10))
    grid = np.zeros((len(maze), len(maze[0])))
    
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == 'S':
                grid[i][j] = 0.2
            elif maze[i][j] == 'G':
                grid[i][j] = 0.4
            elif maze[i][j] == '#':
                grid[i][j] = 0.8
            elif maze[i][j] == '.':
                grid[i][j] = 0.1
            elif maze[i][j] == '^':
                grid[i][j] = 0.3
            elif maze[i][j] == '~':
                grid[i][j] = 0.5
    
    sns.heatmap(grid, cmap='YlOrRd', square=True, cbar=False, linewidths=0.5, linecolor='black')
    ax = plt.gca()
    # Annotate special cells and path
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == 'S':
                ax.text(j + 0.5, i + 0.5, 'S', ha='center', va='center', color='black', fontsize=12, fontweight='bold')
            elif maze[i][j] == 'G':
                ax.text(j + 0.5, i + 0.5, 'G', ha='center', va='center', color='black', fontsize=12, fontweight='bold')

    if path:
        for (i, j) in path:
            if maze[i][j] not in ('S', 'G'):
                ax.text(j + 0.5, i + 0.5, 'X', ha='center', va='center', color='blue')

    plt.title('Maze Visualization')
    plt.show()
```

```python
# Find start and goal positions
def find_positions(maze):
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == 'S':
                start = (i, j)
            elif maze[i][j] == 'G':
                goal = (i, j)
    return start, goal
```

### Generate and visualize maze

```python
maze = generate_maze()
start, goal = find_positions(maze)

print("Initial Maze:")
visualize_maze(maze)
```

## Uninformed Search

Utility

```python
# Get neighbors of a position
def get_neighbors(maze, pos):
    rows, cols = len(maze), len(maze[0])
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # up, down, left, right
    neighbors = []
    
    for d in directions:
        ni, nj = pos[0] + d[0], pos[1] + d[1]
        if 0 <= ni < rows and 0 <= nj < cols and maze[ni][nj] != '#':
            neighbors.append((ni, nj))
    
    return neighbors

results = {}
```

### BFS


```python
def bfs(maze, start, goal):
    queue = deque([(start, [start])])
    visited = set([start])
    nodes_expanded = 0
    max_memory = 0
    
    start_time = time.time()
    
    while queue:
        max_memory = max(max_memory, len(queue))
        pos, path = queue.popleft()
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            cost = sum(TERRAIN_COSTS[maze[i][j]] for (i,j) in path[1:-1])
            return {
                'path': path,
                'cost': cost,
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time
            }
        
        for neighbor in get_neighbors(maze, pos):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    
    return None
```

Showing results for BFS

```python
name = "BFS"
print(f"\nRunning {name}...")
result = bfs(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### DFS

```python
def dfs(maze, start, goal):
    stack = [(start, [start])]
    visited = set([start])
    nodes_expanded = 0
    max_memory = 0
    
    start_time = time.time()
    
    while stack:
        max_memory = max(max_memory, len(stack))
        pos, path = stack.pop()
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            return {
                'path': path,
                'cost': sum(TERRAIN_COSTS[maze[i][j]] for (i, j) in path[1:]),
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time
            }
        
        # Reverse to explore in consistent order (for comparison)
        for neighbor in reversed(get_neighbors(maze, pos)):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append((neighbor, path + [neighbor]))
    
    return None
```

Showing results for DFS

```python
name = "DFS"
print(f"\nRunning {name}...")
result = dfs(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### DLS

```python
def dls(maze, start, goal, limit):
    stack = [(start, [start], 0)]
    visited = set([start])
    nodes_expanded = 0
    max_memory = 0
    complete = True
    
    start_time = time.time()
    
    while stack:
        max_memory = max(max_memory, len(stack))
        pos, path, depth = stack.pop()
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            return {
                'path': path,
                'cost': sum(TERRAIN_COSTS[maze[i][j]] for (i, j) in path[1:]),
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time,
                'complete': True
            }
        
        if depth < limit:
            for neighbor in reversed(get_neighbors(maze, pos)):
                if neighbor not in visited:
                    visited.add(neighbor)
                    stack.append((neighbor, path + [neighbor], depth + 1))
        else:
            complete = False
    
    return {
        'path': None,
        'complete': complete,
        'nodes_expanded': nodes_expanded,
        'max_memory': max_memory,
        'time': time.time() - start_time,
        'cost': "inf"
    }
```

Showing results for DLS

```python
name = "DLS"
print(f"\nRunning {name}...")
result = dls(maze, start, goal, 20)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### UCS

```python
def ucs(maze, start, goal):
    heap = [(0, start, [start])]
    visited = set()
    nodes_expanded = 0
    max_memory = 0
    
    start_time = time.time()
    
    while heap:
        max_memory = max(max_memory, len(heap))
        cost, pos, path = heapq.heappop(heap)
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            return {
                'path': path,
                'cost': cost,
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time
            }
        
        if pos in visited:
            continue
            
        visited.add(pos)
        
        for neighbor in get_neighbors(maze, pos):
            if neighbor not in visited:
                new_cost = cost + TERRAIN_COSTS[maze[neighbor[0]][neighbor[1]]]
                heapq.heappush(heap, (new_cost, neighbor, path + [neighbor]))
    
    return None
```

Showing results for UCS

```python
name = "UCS"
print(f"\nRunning {name}...")
result = ucs(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### IDS

```python
def ids(maze, start, goal):
    depth = 0
    total_nodes = 0
    total_memory = 0
    start_time = time.time()
    
    while True:
        result = dls(maze, start, goal, depth)
        total_nodes += result['nodes_expanded']
        total_memory = max(total_memory, result['max_memory'])
        
        if result['path'] is not None:
            end_time = time.time()
            return {
                'path': result['path'],
                'cost': sum(TERRAIN_COSTS[maze[i][j]] for (i, j) in result['path'][1:]),
                'nodes_expanded': total_nodes,
                'max_memory': total_memory,
                'time': end_time - start_time
            }
        elif result['complete']:
            return None
        
        depth += 1
```

Showing results for IDS

```python
name = "IDS"
print(f"\nRunning {name}...")
result = ids(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

## Informed Search

Utility

```python
# Heuristic function (Manhattan distance)
def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])
```

### Best First Search

```python
def best_first(maze, start, goal):
    heap = [(heuristic(start, goal), start, [start])]
    visited = set()
    nodes_expanded = 0
    max_memory = 0
    
    start_time = time.time()
    
    while heap:
        max_memory = max(max_memory, len(heap))
        _, pos, path = heapq.heappop(heap)
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            return {
                'path': path,
                'cost': sum(TERRAIN_COSTS[maze[i][j]] for (i, j) in path[1:]),
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time
            }
        
        if pos in visited:
            continue
            
        visited.add(pos)
        
        for neighbor in get_neighbors(maze, pos):
            if neighbor not in visited:
                heapq.heappush(heap, (heuristic(neighbor, goal), neighbor, path + [neighbor]))
    
    return None
```

Showing results for Best first search

```python
name = "Best First"
print(f"\nRunning {name}...")
result = best_first(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### A* Search

```python
def a_star(maze, start, goal):
    heap = [(0 + heuristic(start, goal), 0, start, [start])]
    visited = set()
    nodes_expanded = 0
    max_memory = 0
    
    start_time = time.time()
    
    while heap:
        max_memory = max(max_memory, len(heap))
        _, cost, pos, path = heapq.heappop(heap)
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            return {
                'path': path,
                'cost': cost,
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time
            }
        
        if pos in visited:
            continue
            
        visited.add(pos)
        
        for neighbor in get_neighbors(maze, pos):
            if neighbor not in visited:
                new_cost = cost + TERRAIN_COSTS[maze[neighbor[0]][neighbor[1]]]
                heapq.heappush(heap, (new_cost + heuristic(neighbor, goal), new_cost, neighbor, path + [neighbor]))
    
    return None
```

Showing results for A* search

```python
name = "A*"
print(f"\nRunning {name}...")
result = best_first(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### Weighted A* Search

```python
def weighted_a_star(maze, start, goal, weight=1.5):
    heap = [(0 + weight * heuristic(start, goal), 0, start, [start])]
    visited = set()
    nodes_expanded = 0
    max_memory = 0
    
    start_time = time.time()
    
    while heap:
        max_memory = max(max_memory, len(heap))
        _, cost, pos, path = heapq.heappop(heap)
        nodes_expanded += 1
        
        if pos == goal:
            end_time = time.time()
            return {
                'path': path,
                'cost': cost,
                'nodes_expanded': nodes_expanded,
                'max_memory': max_memory,
                'time': end_time - start_time
            }
        
        if pos in visited:
            continue
            
        visited.add(pos)
        
        for neighbor in get_neighbors(maze, pos):
            if neighbor not in visited:
                new_cost = cost + TERRAIN_COSTS[maze[neighbor[0]][neighbor[1]]]
                heapq.heappush(heap, (new_cost + weight * heuristic(neighbor, goal), new_cost, neighbor, path + [neighbor]))
    
    return None
```

Showing results for weighted A* search

```python
name = "Weighted A*"
print(f"\nRunning {name}...")
result = best_first(maze, start, goal)
results[name] = result

if result and 'path' in result:
    print(f"{name} found a path with cost {result['cost']}")
    print(f"Nodes expanded: {result['nodes_expanded']}")
    print(f"Max memory usage: {result['max_memory']}")
    print(f"Time taken: {result['time']:.4f} seconds")
    visualize_maze(maze, result['path'])
else:
    print(f"{name} did not find a path")
```

### Function for generating heat map

```python
def generate_heatmap(results):
    algorithms = list(results.keys())
    metrics = ['Completeness', 'Optimality', 'Memory', 'Time', 'Nodes Expanded']
    
    # Normalize metrics for heatmap
    max_memory = max(r['max_memory'] for r in results.values())
    max_time = max(r['time'] for r in results.values())
    max_nodes = max(r['nodes_expanded'] for r in results.values())
    
    data = []
    for algo in algorithms:
        res = results[algo]
        row = [
            1 if res['path'] else 0,  # Completeness
            1 if algo in ['UCS', 'A*'] else 0,  # Optimality
            res['max_memory'] / max_memory,
            res['time'] / max_time,
            res['nodes_expanded'] / max_nodes
        ]
        data.append(row)
    
    plt.figure(figsize=(12, 6))
    sns.heatmap(data, annot=True, xticklabels=metrics, yticklabels=algorithms, cmap='YlGnBu')
    plt.title('Search Algorithms Performance Comparison')
    plt.show()
```

### Generating Heatmap

```python
generate_heatmap(results)
```
